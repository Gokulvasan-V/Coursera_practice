### Analysis of multivariate data - NHANES case study

```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
from scipy import stats
```

```python
# Download NHANES 2015-2016 data
da = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/dataset/NHANES.csv")
da.head() # here we didn't give Max column option
```

#### Quantitative bivariate data

```python
sns.regplot(x="BMXLEG", y="BMXARML", data=da, fit_reg=False, scatter_kws={"alpha": 0.2})
```

```python
#ignore deprecation warning message

#sns.jointplot(x="BMXLEG", y="BMXARML", kind='kde', data=da).annotate(stats.pearsonr) # here annotate function shows error

sns.jointplot(x="BMXLEG", y="BMXARML", kind='kde', data=da,
              shade=True 
             )
```

```python
#ignore deprecation warning message

#sns.jointplot(x="BPXSY1", y="BPXDI1", kind='kde', data=da).annotate(stats.pearsonr) # here annotate function shows error

sns.jointplot(x="BPXSY1", y="BPXDI1", kind='kde', data=da, shade=True )
```
